Basic File Operations (Text Files)

When working with files in Python, the golden rule is to use the with statement (also known as a context manager). This ensures that the file is automatically closed when you are done with it, even if an error occurs during your script.

Here are the most common file modes you need to know:



```
Mode,Description
'r',Read (Default). Opens a file for reading; throws an error if it doesn't exist.
'w',"Write. Opens a file for writing; creates it if it doesn't exist, and overwrites it if it does."
'a',Append. Opens a file to add data to the end; creates it if it doesn't exist.
```



Here is a script demonstrating how to write to a file and then read from it:



```
# script_file_ops.py

# 1. Writing to a file (creates the file or overwrites if it exists)
def write_log_file():
    # The 'w' mode opens the file for writing
    with open("server_logs.txt", "w") as file:
        file.write("INFO: Server started successfully.\n")
        file.write("WARNING: High memory usage detected.\n")
        file.write("ERROR: Failed to connect to database.\n")
    print("Log file created and populated.")

# 2. Appending to an existing file
def append_to_file():
    # The 'a' mode appends to the end without erasing previous content
    with open("server_logs.txt", "a") as file:
        file.write("INFO: Database reconnected.\n")
    print("New log entry appended.")

# 3. Reading from a file
def read_log_file():
    print("\n--- Reading Current Log File ---")
    # The 'r' mode is default, but explicitly stating it is good practice
    with open("server_logs.txt", "r") as file:
        lines = file.readlines() # Reads all lines into a list
        
        for line in lines:
            # .strip() removes the invisible newline characters (\n) at the end
            print(line.strip())

# Execute the functions
write_log_file()
append_to_file()
read_log_file()
```
Data Processing (CSV to JSON)

In the real world, you rarely process plain text; you process structured data like CSVs or JSON files.

Imagine you have exported a list of servers and their CPU usage to a CSV file. Your assignment is to read this CSV, filter out only the servers that are dangerously overloaded (CPU > 85%), and save that filtered list as a JSON file for another application to consume.

Step 1: The Input Data (servers.csv)

Create a mock file in the same directory as your script with this data:



```
hostname,ip_address,cpu_usage
web-server-01,192.168.1.10,45
web-server-02,192.168.1.11,92
db-server-01,192.168.1.50,30
api-server-01,192.168.1.60,88
```

Step 2: The Data Processing Script



```
# script_data_processing.py
import csv
import json

def process_server_data(input_csv, output_json):
    overloaded_servers = []
    cpu_threshold = 85

    # 1. READ the CSV file
    with open(input_csv, "r") as csv_file:
        # DictReader automatically uses the first row as dictionary keys
        reader = csv.DictReader(csv_file)
        
        for row in reader:
            # 2. PROCESS the data (filter based on CPU usage)
            # CSV data is read as strings, so we must convert cpu_usage to an integer
            if int(row["cpu_usage"]) > cpu_threshold:
                # Add a new status key/value pair to the dictionary
                row["status"] = "CRITICAL"
                overloaded_servers.append(row)

    # 3. WRITE the processed data to a JSON file
    with open(output_json, "w") as json_file:
        # json.dump writes the Python list of dictionaries directly to a JSON file
        # indent=4 makes the output file highly readable
        json.dump(overloaded_servers, json_file, indent=4)

    print(f"Processing complete! Found {len(overloaded_servers)} overloaded servers.")
    print(f"Results saved to {output_json}")

# Execute the function
process_server_data("servers.csv", "alerts.json")
```
If you run this, it will generate an alerts.json file containing only web-server-02 and api-server-01, perfectly formatted.

Step 2: Write the Web Scraper

Here is a complete, commented Python script. It will visit the website, extract the text of each quote along with its author, print the results to your screen, and finally save them into a CSV file (building perfectly on the file operations you just learned!).


Create a file named simple_scraper.py and paste the following code:

```
import requests
from bs4 import BeautifulSoup
import csv

def scrape_quotes():
    # 1. Define the target URL
    url = "http://quotes.toscrape.com/"
    print(f"Connecting to {url}...")

    # 2. Send an HTTP GET request to the website
    response = requests.get(url)

    # 3. Check if the request was successful (Status Code 200 means OK)
    if response.status_code != 200:
        print(f"Failed to retrieve webpage. Status code: {response.status_code}")
        return

    print("Successfully connected! Parsing data...\n")

    # 4. Parse the raw HTML using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # 5. Find the specific HTML elements containing our data
    # On this specific website, quotes are inside <span class="text">
    # and authors are inside <small class="author">
    quote_elements = soup.find_all('span', class_='text')
    author_elements = soup.find_all('small', class_='author')

    # 6. Extract the text and store it
    extracted_data = []
    
    # We loop through both lists simultaneously
    for i in range(len(quote_elements)):
        # .text removes the HTML tags (like <span>) and just leaves the words
        quote_text = quote_elements[i].text
        author_name = author_elements[i].text
        
        extracted_data.append([author_name, quote_text])
        print(f"Extracted: {author_name} - {quote_text[:30]}...") # Print a preview

    # 7. Save the extracted data to a CSV file
    save_to_csv(extracted_data)

def save_to_csv(data):
    # Use the 'w' mode to write a new file, specifying utf-8 for special characters
    with open('scraped_quotes.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # Write the header row first
        writer.writerow(['Author', 'Quote'])
        # Write all the data rows
        writer.writerows(data)
    
    print("\nSuccess! All data has been saved to 'scraped_quotes.csv'.")

# Execute the script
if __name__ == "__main__":
    scrape_quotes()
```

Step 3: How It Works (The Core Concepts)

To understand this script for your assignment, here is a breakdown of the core mechanics:



*   requests.get(url): This simulates your web browser typing in the URL and hitting Enter. It downloads the raw source code of the webpage.

*   BeautifulSoup(...): Raw HTML is just a giant, messy string of text. BeautifulSoup turns that messy string into a navigable tree structure.
*  soup.find_all(...): This is the most important part of scraping. You have to inspect the target website's HTML (usually by right-clicking on the webpage and selecting "Inspect") to find the specific HTML tags and class names that wrap the data you want. We told BeautifulSoup to find every span tag that has a class of text.
